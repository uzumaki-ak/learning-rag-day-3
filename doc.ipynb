```python
import json
from langchain.docstore.document import Document

def load_json(path:str):
     with open(path,'r',encoding='utf-8') as f:
          data=json.load(f)

          documents =[]
          for entry in data:
               page_content = f"Title: {entry.get('title','')}\n\ncontent:{entry.get('content','')}"
               metadata = {
                    "title":entry.get('title',''),
                    "source": path
                }
               documents.append(Document(page_content=page_content,metadata=metadata))
          return documents
```

```python
load_json('xyz.json')
```

```python
SimpleDirectoryReader(input_files=['xyz.json'],encoding="utf-8").load_data()
```
